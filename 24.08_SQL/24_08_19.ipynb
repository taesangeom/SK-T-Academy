```python
# 필요한 기능들에 대해서 함수화를 최대한 시도
# -1명 학생에 대한 정보를 생성하는 함수
# +1명 학생에 대한 정보를 바탕으로 총점을 구하는 함수
# +1명 학생에 " 평균
# +1명 학생 " 출력(이름,총점,평균) 함수
```

```python
# - 1명 학생에 대한 정보를 생성하는 함수
# 입력 : 이름, 국어, 수학, 영어, 과학
# 기능 : 학생 정보에 대한 Dict 생성
# 출력 : 1명 학생에 대한 정보 dict
def create_student( name, kor, math, eng, sci):
    return {"name":name, "korean":kor, "math":math,
            "english":eng, "science":sci}

# + 1명 학생에 대한 총점을 계산해주는 함수
# 입력 : 1명 학생의 정보
# 기능 : 총점 께산
#출력 : 총점

def student_get_sum(student):
    score_sum=student["korean"]+student["math"]+student["english"]+student["science"]
    return score_sum


def student_get_avg(student):
    score_sum=student["korean"]+student["math"]+student["english"]+student["science"]
    n=len(student.keys())-1
    return score_sum/n

#출력 함수 : 이름,총점,평균

def student_to_string(student):
    temp="{0}\t{1}\t{2}"
    return temp.format(student["name"],student_get_sum(student),student_get_avg(student))
```

```python
#####  일반적으로 프로그램의 코드들을 작성하는 스타일과 조금 달라요!!!
#      ==> 객체지향으로 설계 : Class --->  Object
# : 필요한 기능 중심으로, 조립을 하는 방식
# : 많은 시행착오, 경험, 노하우, 트렌드.....
# :참고) 데이터분석/ 모델링쪽은 : 아주 심하는지는 않음!!!
#                                 기본적인 부분 // 약간의 스타일들이 있어서..
# :참고) kaggle 오픈 사이트 : 올려둔 샘플코들들이 클래스를 사용하는 경우!!!
#                             이해하기 위해서도 알기는 해야함!!!!


# 클래스:class : 속성 + 기능( 뭔가 틀을 생성하는것이 class)
#                점수, 이름 : 속성 ---> property
#                평균구해주세요, 총점계산해주세요 : 기능 --> method

# 객체 : 클래스를 기반으로 생성된 구체적인 대상!!!!!
#        붕어빵틀 : 클래스 ---> 구체적으로 만들어진 붕어빵1개 : 객체
# self : 자기 자신을!!!!!!!
```

```python
class Student:
# 생성자 : 필요한 초기 속성들에 대한 초기화!!!!
#         ==> 파이썬에서 클래스 생성시 초기화  미리 만든 함수 특별 Method : __(양옆2개)__
    def __init__(self,name,korean,math,english,science):  #생성자를 def __init__로 정의하는구나 / self = this / private 변수 정의도 필요없고, 생성자를 통해 언급?
        self.name=name
        self.korean=korean
        self.math=math
        self.english=english
        self.science=science

    def get_sum(self):
        return self.korean+self.math+self.english+self.science

    def get_avg(self):
        return self.get_sum()/4

    def student_to_string(self):
        temp="{0}  {1}  {2:.1f}".format(self.name,self.get_sum(),self.get_avg())
        return temp

    def __eq__(self,value):
        return self.get_sum()==value.get_sum()

    def __ne__(self,value):
        return self.get_sum()!=value.get_sum()

    def __gt__(self,value): #great
        return self.get_sum()>value.get_sum()

    def __ge__(self,value): #great equal
        return self.get_sum()>=value.get_sum()

    def __le__(self, value): #less equal
        return self.get_sum()<=value.get_sum()

    def __lt__(self, value): #less
        return self.get_sum()<value.get_sum()
```

```python
# 정사각형 관련 클래스
class Square:
    def __init__(self, size):
        self.size = size
    def get_area(self):
        return self.size * self.size
```

```python
s1 = Square(10)
s1
```

```python
s2=Square(20)
s2.size=30 #외부접근 가능하면 값 바꿀 수도 있는 문제점
```

```python
s2.get_area()
```

```python
#클래스 내부의 속성값에 대한 외부접근 막기 위해서 속성값앞에만  __를 이용해서 private 지정

class Square:
    def __init__(self, size):
        self.__size = size
    def get_area(self):
        return self.__size * self.__size
```

```python
#내부 속성값 접근에 대한 권한을 공식적으로 요청해라. 함수를 통해서 기능적으로 내부에 접근할 수 있듯이(공식통로)

class Square:
    def __init__(self, size):  # __변수는 private 변수이며, getter, setter 통해 접근한다 (java 처럼)
        self.__size = size
    def get_area(self):
        return self.__size * self.__size
    def get_size(self):
        return self.__size
    def set_size(self,size):
        self.__size=size
```

```python
s1=Square(10)
s1.get_area()
```

```python
s1.set_size(30)
s1.get_area()
```
