```python
# 필요한 패키지들
import pandas as pd
import urllib.request # <-- http 통신을 요청하려고 함...
import json           # <-- 요청한 json 양식 처리 편의를 위해서..
```

```python
# Step1) 내가 필요한 정보들을 요청할 주소 생성!!!!!
# --> 메뉴얼 잘 봐야 함!!!
#     메뉴얼 상에서 주소 요청하는 방식 get 방식으로 함!!!
#     : 요청할 사항들에 대해서 주소로 직접 작성해서 요청!!!!
#       baseurl?요청할파라미터=요청할값&~~~~~~~~~~~
# --> 참고 일반적인 사이트들은 이 요청 파라미터들을 다 뒤져야한다.
```

```python
# 기본 주소
url_p1 = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"

#필수항목 : key 값
key="bf30b4505678332619c7114435344b4d"
#부가적인 요청사항이 있다면 : itemPerPage= 40으로 갯수 늘려서 요청

url_p2="40"
#최종적으로 내가 완성할 주소를 요청
url=url_p1+"?key="+key+"&itemPerPage="+url_p2
url
```

```python
#참고 : 내가 40개를 요청한다고 해도 꼭 40개가 오는 것은 아니다.
# 꼭 다른 데이터의 수를 체크
```

```python
tot_list=[]
for movie in movie_data['movieListResult']['movieList']:
        cd=movie['movieCd']
        nm=movie['movieNm']
        en=movie['movieNmEn']
        year=movie['openDt']
        if len(movie['directors'])==0:
            direct=""
        else:
            direct=movie['directors'][0]['peopleNm']
        #print(cd,nm, year, direct) # 프린트 대신 리스트 혹은 디큐에 잘 넣자 -> 모아서 -> dataframe으로 변환하면 됨
        tot_list.append([cd,nm,en, year, direct])
tot_list

```

```python
# ==> 위에 과정은 개별 정보들에 대해서 접근 OK
# ==> 이 정보들은 아직 어딘가에 담겨있는 것이 아니라 단순 출력!!
#     이 정보들을 잘 담아두자!!!!!!
#     DF에서 직접 해도 되지만,,,,,무거워서,,,시간 많이 걸림!!!
#     현실적으로 이 정보들을 list/deque / dict etc 담은 후
#     나중에 한 번에 DF으로 변환
```

```python
movie_df=pd.DataFrame(data=tot_list,columns=["movidCd","movieNm","movieEnNm","opemDt","director"])
movie_df
```

```python
# 할 일
# 1. 개봉일에 대한 정보를 바탕으로 년도 컬럼 추가
#                                  월   컬럼 추가
#                                  일   컬럼 추가
# 2. movieCd 컬럼의 값이 메뉴얼상의 코드값인거 같음!!!!
#    ==> 코드값이니 진짜로 유니크한지 체크!!!!!!!!
#    ==> 혹 이 값들이 유니크하면,,,,,DF의 가로줄 index으로 사용가능하니까
```

```python
movie_df["year"]=movie_df.apply(lambda x:x[3][:4],axis=1)
movie_df["month"]=movie_df.apply(lambda x:x[3][4:6],axis=1)
movie_df["day"]=movie_df.apply(lambda x:x[3][6:],axis=1)
movie_df
```

```python
movie_df["year"] = movie_df.loc[:, "openDay"].apply( lambda x: x[:4])
movie_df["month"] = movie_df.loc[:, "openDay"].apply( lambda x: x[4:6])
movie_df["day"] = movie_df.loc[:, "openDay"].apply( lambda x: x[6:])
movie_df.head() # 이렇게도 먼저 loc으로 접근해서도 처리 가능
```
