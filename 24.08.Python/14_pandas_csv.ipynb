```python
#read_csv()중요 옵션 1) sep
path='/content/01_olive.csv'
data=pd.read_csv(path,sep=",")
data.head()
```

```python
#read_csv()중요 옵션 2) header
#->csv 파일의 첫번째 줄을 header -> 컬럼 이름으로 사용할지 말지
path='/content/01_olive.csv'
data=pd.read_csv(path,sep=",",header=None)  #헤더가 파일로 내려갈 수도 있음
data.head()
```

```python
# read_csv 중요 옵션3) skiprows
# ==> csv 파일의 몇 번 쨰 줄을 넘어서 읽기 시작할지
#     : 몇 번째 줄부터 읽을까!!!!!
# ==> csv 파일 앞에,,,,라이센스,법적사항,,,,소유권,,,,,
path = "/content/01_olive.csv"
data = pd.read_csv(path, sep=",", header=None,
                   skiprows=2) # 헤더제외하고 0,1번째 줄 스킵하고 두번째 부터 읽겠다
data.head()
```
