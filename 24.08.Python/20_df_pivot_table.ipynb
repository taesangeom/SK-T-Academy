```python
import pandas as pd
import numpy as np
```

```python
# 본인 환경에 맞는 11.slaes_funnel.xlsx 파일 불러오세요..
#colab이여서 그냥 간단히 드래그로 업로드!!!!
# ==> 어느 시트인지 체크해서 불러오셔야 함!!!!
path = '/content/11_sales-funnel.xlsx'
data = pd.read_excel( path, sheet_name ="Sheet1")
data.head()
```

```python
data.info()
```

```python
#매출 관련 데이터로 인사평가를 해보자
# -> 본인이 인사팀장의 관점으로 바라보자

#데이터에 대한 간략한 설명

# 데이터에 대한 간략한 설명
# Account : 고객의 계좌번호
# Name : 고객의 이름
# Rep  : 영업 담당자( 실무 영업사원 )
# Manager : 영업 담당자의 팀장
# Product : 판매한 물건
# Quantity : 판매한 물건의 수량
# Price    : 판매한 물건의 가격
# Status   : 현재 진행 상황
```

```python
# pivot_table을 활용해서 새롭게 판을 짜는 방식!!!!
# 1. 가로에 무엇을 볼지
# 2. 세로에 무엇을 볼지
# 3. 1/2에 의해서 만들어진 공간을 어떻게 채울까!!!!
#    ==> 여러개의 원본 데이터들을 어떻게 집계처리해서 칸 대표화!!!

# pandas 2.x
# --> 예전에는 대충, 가로만 세팅해도 알아서 세로하고, 값들 처리
# --> FM : 가로 세팅, 세로 세팅, 무엇을 볼지 명시!!!!
```

```python
# 참고) 지금 pandas 2.x에서는 안 돌아가는 코드
pd.pivot_table( data, index=["Manager"])
```

```python
#q) 팀장들에 대해서 팀에 속한 매출액을 보고 싶다.
data["income"]=data["Price"]*data["Quantity"]
```

```python
data.head()
```

```python
data['income'].groupby(data["Manager"]).sum()
```

```python
pd.pivot_table(data,index=['Manager'],values=['income'],aggfunc=['sum'])
```

```python
pd.pivot_table(data,index=['Manager','Rep'],values=['income'],aggfunc=['sum'])
```

```python
pd.pivot_table(data,index=['Manager'],values=['income'])  #평균인가?
```

```python
len(data['Manager'].unique())
```

```python
# 또는 pivot table 이용

# Q) 팀장들에 대해서 팀에 속한 매출액을 보고 싶다!!
# --> 평가의 기준이 오로지 매출
# --> 값들에 대해서 특별하게 지정하지 않으면 : 평균!!
pd.pivot_table(data, index=["Manager"], values=["Price"] ) # 가로 매니저, 세로, 가격
```

```python
# FM적으로 pivot_table에 대한 속성값
# index : 가로줄에 대한 세팅 --> 보려는 기준!!!
# values : 세로줄에 대한 세팅 --> 보려는 속성!!
# aggfunc : 어떻게 여러 값을 집계해서 채울까!!!!!
# + columns : 속성별로 나눠서 보자!!!!분류용!!!!
pd.pivot_table(data, index=["Manager"], values=["Price"],aggfunc=["sum"] )
```

```python
pd.pivot_table(data,index=['Manager','Rep'],values=['Price','Quantity'],columns=['Product'],aggfunc=['sum'],fill_value=0,margins=True)
# 걍 개복잡함
```

```python
### ++ 집계 처리를 하고자하는 방식이
#      보려는 속성/항목별로 다르게 하고 싶다!!!
# ==> price : 평균, 총계
# ==> quatnity : 총계
# --------------> Dict로 쌍에 대한 구조 표현!!!!
```

```python
pd.pivot_table( data, index=["Manager","Rep"], # group by
               values =["Price", "Quantity"], # --> 일반적인 수치형컬럼  #Select
                columns = ["Product"],# --> 카테고리형 컬럼  # 세부항목
                aggfunc={
                    "Price":["mean","sum"],
                    "Quantity":["sum"]
                },
                fill_value=0,
                )
```

```python
### pivot_table 의 목적 ###
# ===> 수집된 데이터를 내가 보고자 하는 "항목 중심"으로 바라보자!!!
#      가로 : index
#      세로 : values + 항목별 쪼개서 : columns
#      --> aggfunc : 새로운 판에 칸을 무엇을 채울지!!!
# + 기존에 있는 항목이 아니라
#   새롭게 본인이 직접 만든 컬럼에 대해서도 생성하고 하시면 됨!
# ==> 수집한 데이터의 여러 속성/ 탐색!!!!!!!!!
```
