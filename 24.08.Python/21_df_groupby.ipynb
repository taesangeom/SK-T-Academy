```python
### Group by ###
# --> sql에서 하셨던 그 group by와 동일함!@!!!
# --> 기존의 데이터들을 어떻게 묶어서 볼지에 대한 새로운 기준!!!
#     & 여러개의 값들을 묶어서 표현 :집계처리(대표값!!!)
# sql) select avg(zumsu) from data group by class;
# pandas) df.group_by(by="class").agg(집계처리)
#         df.group_by(by="class").집계함수
```

```python
import pandas as pd
import numpy as np
```

```python
path='/content/10_phone_data.csv'
data=pd.read_csv(path,sep=",")
data.head()
```

```python
data.info()
```

```python
data.loc[:,"item"].value_counts()
# item 컬럼의 유니크한 값들의 종류와 갯수를
# 갯수에 대한 자동 정렬
```

```python
#duration 컬럼의 최대 사용시간은 얼마인가요
data.loc[:,"duration"].max()
```

```python
data.describe()
# 수치형 컬럼들에 대해 기본적인 정보를 줌, 최대,최소,사분위수
```

```python
# Q) item 항목들 중에서 전화인 call인 경우에 대해서
#    ==> 사용한 시간의 총합을 구해보세요!
#data.loc[data.loc[:,'item']=='call'] 자체가 from 서브쿼리의 역할
# 불리언 인덱싱으로 가자

data.loc[data.loc[:,'item']=='call'].loc[:,'duration'].sum()
```

```python
#month에 대한 종류와 건수 체크

data.loc[:,'month'].value_counts()
```

```python
# 주어진 데이터들을 월별값을 기준으로 묶어서 보자
data.groupby(by='month').head()  # unique month 별로 5개 씩 걸림
# 눈으로 안보이지만, DF 으로 생각을 하면 된다.
```

```python
data.groupby(by="month").groups   # 딕셔너리로 관리중이란걸 알 수 있다.
```

```python
data.groupby(by="month").groups.keys()
```

```python
data.loc[:,'month'].unique()
```

```python
#month 별로 pivot_tabel 이용해서 평균
pd.pivot_table(data,index=['month'],values=['duration'],aggfunc=["mean"]) # type은 DataFrame 이다
```

```python
data.groupby(by='month').loc[:,'duration'].mean()
```

```python
data.groupby(by='month')['duration'].mean()  #type는 Series이다 1차원
```

```python
#item 컬럼의 값중에서 종류가 call 인 경우에 한해서 (전화통화) network 통신사별로 duration 총합을 구해보자
data.loc[data.loc[:,'item']=='call',:].groupby(by='network')["duration"].sum()
```

```python
# 참고, 묶는 기준이 여러개일 수 있음 -> Manager /Rep , Month/ item 몇건인지 보자
data.groupby(by=['month','item'])['date'].count()
```

```python
data.groupby(by=['month','item']).agg(
    {
        "date":"first",
        "duration":"sum"
    }
)
```
