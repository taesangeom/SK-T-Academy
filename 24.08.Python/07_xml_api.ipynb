```python
import pandas as pd
import urllib.request
#xml 태그중심 html : 태그중심의 언어들로 표현한 것들 :beautifulSOUP
from bs4 import BeautifulSoup
```

```python
# Step1) xml 양식으로 지금 시점에서
#        최신영화 40개를 요청하는 url을 생성
base_url="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml"
key="bf30b4505678332619c7114435344b4d"
itemPerPage="40"
url=base_url+"?key="+key+"&itemPerPage="+itemPerPage
res=urllib.request.urlopen(url)
# xml 양식으로 요청한 정보들 처리 위해서는 / bs -> parser: "html.parser", "lxml.parser"
# 보통 html.parser 사용해도 됨
#     xml의 사이즈가 크거나 하면, 별도의 전문 큰사이즈 파서 사요으...
#     메뉴얼을 참조하시면 됨!!!!
soup=BeautifulSoup(res,"html.parser")

```

```python
# ==> i번째 영화에 대한 정보를 기준!!!
idx = 0 # 0~~~~끝까지 숫자로 돌리면 됨!!!!
i_movie = soup.find_all("movie")[idx]
print(i_movie.find("moviecd").text)
print(i_movie.find("movienm").text)
print(i_movie.find("movienmen").text)
print(i_movie.find("opendt").text)
print(i_movie.find_all("directors")[0].find("director").find("peoplenm").text)
```

```python
mv_dict=[]
for idx in range(len(soup.find_all("movie"))):
    i_movie = soup.find_all("movie")[idx]
    moviecd=i_movie.find("moviecd").text
    movienm=i_movie.find("movienm").text
    movienmen=i_movie.find("movienmen").text
    opendt=i_movie.find("opendt").text
    if len(i_movie.find_all("peoplenm"))==0:
        director="감독없음"
    else:
        director=i_movie.find("peoplenm").text
    temp={
        "moviecd":moviecd,
        "movienm":movienm,
        "movienmen":movienmen,
        "opendt":opendt,
        "director":director
    }
    mv_dict.append(temp)
print(mv_dict)
```
