```python
# pandas에서 결측치에 대한 처리!!!
# ==> pandas가 결측치로 인정한 경우에 대해서만!!!
# 빵구난 데이터가 nan, "", None ;;; ==>  처리를 유의!!!!
# ==> 아래서 기준은 이 중에서 nan에 대해서 한정!!!!
```

```python
# ==> 빵구난 데이터 처리들이 중요한 이슈!!
#     EDA를 기반으로 누락된 데이터들을 어떻게 할지 결정!!!
#     : ML 모델을 돌리기 위해서는 코드적을 처리를 안 하면 에러!!!
# ==> 처리방법) 빵구난 데이터를 지우자 : dropna
#                                        문제는 경우에 따라서 데이터가 많이 줄어듬!!
#               빵구난 데이터를 채우자 : 정답은 없음!!!!
#                                        분석자가 주관적으로 해야함!!
#                                        1개 대표값 : 평균, 중앙값, 최대값 etc
#                                        유사한 데이터를 보고 추정!!!
#                                        모델을 만들어서 예측값으로 채우는 방법..
#                                        fillna + lambda apply

# 참고) 특정한 줄(가로, 세로) 지울 때 : drop
#       nan인 데이터를 지울 때        : dropna
```

```python
# kaggle/정제된 데이터 셋 & 실제 :
# ==> 결측치에 대한 처리 여부!!! 성능에 좌우를 함!!!!
```

```python
from numpy import nan as NA
import pandas as pd
```

```python
data = pd.DataFrame(
    data = [
        [1,6.5, 3],
        [1, NA,NA],
        [NA,NA,NA],
        [NA, 6.5,3]
    ]
)
data
```

```python
data.dropna()
#가로줄 기준으로 모든 정보들이 정상적으로 존재하는 행만 남김
#(모든 컬럼에 대해서 Nan 값이 있는 레코드는 다 제거 ( 빈 문자열 등은 안걸러짐 ))
#장점: 샘플이 모든 속성 값을 다 가지고 있음 (빈 문자열은 남아있을 수 있음)
# 단점 : 샘플 수가 확 줄어들 수 있다.
```

```python
data.dropna(axis=1) # 세로로도 지워버림
```

```python
# dropna에서
# axis : 가로줄/세로줄로 데이터들이 다 있는지 체크!!!
#        axis=0/1
# thresh : 정상적인 데이터가 살아 있는 기준의 수!!!
# how = "all" : 그 줄에 모든 값이 누락된 경우만 제거해주세요!!!
```

```python
data.dropna(axis=0,how="all") # 가로 기준 모든 값이 na인 경우만 지움
```

```python
data.dropna(axis=0,thresh=1) #가로 기준 정상적인 값이 1개 이상이면 살린다.
```

```python
data.dropna(axis=0,thresh=2) #가로 기준 정상적인 값이 2개 이상이면 살린다.
```

```python
# 채우자 fillna ==> nan로 되어 있는 값을 채우자!!!!
# --> 컬럼별로 누락된 값을 각각 채우고 싶다!!
# 예) 1번 컬럼의 누락데이터는 기본값을 0으로 채우고
#     2번 컬럼의 누락데이터는 기본값을 9999로 채우고 싶다!!!
# 참고) 여기서는 기능적인 부분이여서 특정한 값으로 채우는
#       경우만 이야기를 하겠습니다.
```

```python
data.fillna(100) # 100으로 채움
```

```python
data.fillna({0:0,1:999})
#어느 컬럼에 따라 결측치를 다르게 처리 -> dict 쌍 구조 {컬럼:채울 값}
```
