```python
# 목적1 : 앞에서 한 kobis api에서 영화목록에 대한 요청을 하고
#         DF으로 정리하는 과정에서 있어서 : 리스트 대신 Dict
#         + 일단 있는 정보 다 가지고 처리하려고 해서!!!!!

# 목적2 : requests 패키지 활용!!!
#         ==> json 변환에 대한 모듈이 이미 내장되어 있음!!!
#             import json 생략가능함!!!
```

```python
# 필요한 모듈
import pandas as pd
import requests # --> col
# import json : 생략가능함
```

```python
# 요청하려는 api의 주소 작성...
# 1) base url
base_url =" http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
# 2) 실제 요청 파라미터에 대해서 dict로 표현!!!
#    api상에서 사용할 수 있는 요청인터페이스 : key(문자열)
#                             요청인자에 대한 값 : value(숫자, 문자.)
# --> key, itemPerPage
my_params = {
    "key" : "bf30b4505678332619c7114435344b4d",
    "itemPerPage" : 40
}
# 3) 해당하는 요청에 대한
res = requests.get(base_url, params=my_params)

# 4) 중간 FM으로 http 통신에 요청을 하면,,,http 응답 코드 200--> OK
#    https://developer.mozilla.org/en-US/docs/Web/HTTP/Status#successful_responses
if res.ok:
    result=res.json()
else:
    print("실패")
```

```python
directors_list=[]
for idx,data in enumerate(result['movieListResult']['movieList']):
    directors_list.append(data['directors'])
directors_list
```

```python
# Dict 에서 병합!!!!!
d1 = {"A":1234, "B":5678}
d2 = {"B":9999, "C":1000}
{**d1,**d2}  # d2꺼를 겹치면 d2걸로 갱신하고,안 겹치면 생성
```

```python
# * 기호 본 경험
# 1) *args : 파이썬 sum_random(*args): --> 리스트
# 2) **args : def aa(p="12", cc=123)
#       dict = {cc:999, p="34"}
# ===> k/v
```

```python
# --> 전체 영화에 대한 정보들을 담을 리스트...
new_list = []
# idx : 정수 인덱스 : 감독 정보 처리...
# data : 개별 영화에 대한 정보 : dict
for idx, data in enumerate(result["movieListResult"]["movieList"] ):
    one_data = data
    if len(directors_list[idx]) != 0:
        one_data = {**one_data, **directors_list[idx][0]}
        # {"movieCd":1234, "movieNm":"광해"}
        # {"peopleNm":"AAA"}
        # {"movieCd":1234, "movieNm":"광해","peopleNm":"AAA"}
    new_list.append( one_data)
print(new_list)
```

```python
url="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
my_params = {
    "key" : "bf30b4505678332619c7114435344b4d",
    "itemPerPage":20,
    "movieTypeCd" : "220102"

}
res=requests.get(url,params=my_params)
result=res.json()
result
```

```python
sht_list=[]
for i in range(len(result['movieListResult']['movieList'])):
    sht_list.append(result['movieListResult']['movieList'][i]['movieCd'])
sht_list
```

```python

res_list=[]
not_list=[]
for i in sht_list:
    b_url="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json"
    key="bf30b4505678332619c7114435344b4d"
    mvcd=str(i)
    url=b_url+"?key="+key+"&movieCd="+mvcd
    res=requests.get(url)
    result=res.json()
    if 'movieInfoResult' not in result.keys():
        not_list.append(i)
        continue
    data=result['movieInfoResult']['movieInfo']
    mov_cd=data['movieCd']
    mov_nm=data['movieNm']
    if len(data['directors'])==0:
        direct="감독없음"
    else:
        direct=data['directors'][0]['peopleNm']
    if len(data['actors'])==0:
        actor="배우없음"
    else:
        actor=data['actors'][0]['peopleNm']
    if len(data['genres'])==0:
        genre="장르없음"
    else:
        genre=data['genres'][0]['genreNm']
    show_time=data['showTm']

    temp={"영화코드":mov_cd,"영화제목":mov_nm,"감독":direct,"배우":actor,"장르":genre,"상영시간":show_time}
    res_list.append(temp)
res_list




```
